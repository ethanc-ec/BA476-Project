```python
threshold = 0.5
fc = df_nodupe.corr(numeric_only=True)

# Check abs(corr) > threshold and corr != 1 or close to 1,
# then remove NANs and fill remaining NANs w/ 1 since it's just self correlation
fc = (
    fc[((fc > threshold) | (fc < -threshold)) & (fc != 1.0) & (fc < 0.99)]
    .dropna(axis=0, how="all")
    .dropna(axis=1, how="all")
    .fillna(1)
)

sns.heatmap(fc, cmap=spotify_palette, vmin=-1, vmax=1, annot=True, annot_kws={"color": "black"})
plt.title("Correlation Heatmap of Highly Correlated Variables")
plt.show()
```

```python
no_writer = df[df["track_genre"] != "songwriter"]
df_merged = no_writer.merge(no_writer, on="track_id")

mask = df_merged["track_genre_x"] > df_merged["track_genre_y"]

counts = (
    df_merged[mask]
    .groupby(["track_genre_x", "track_genre_y"])
    .agg(collaboration_count=("track_id", "count"))
)

counts.sort_values(by=["collaboration_count"], ascending=[False]).head(20)
```

```python
sorted_counts = counts.sort_values(by=["collaboration_count"], ascending=[False])

order = []
for item in sorted_counts.index.get_level_values(0):
    if item not in order:
        order.append(item)

order_nums = {val: idx for idx, val in enumerate(order)}
counts["order"] = counts.index.get_level_values(0).map(order_nums)
counts.sort_values(by=["order", "collaboration_count"], ascending=[True, False]).head(20)
```
